```python
#Graph of PerCo for GC all veg
nonveg = ['BARE', 'DUFF', 'DOM', 'LI', 'LITR', 'ALG', 'ALGA', 'ALGB', 'ALGG','HAY', 'OPWA', 'H2O', 'MF',"WATER","LI","WRAC"]
veg_gc = data[~data.SPP.isin(nonveg)]
denom = data.groupby("Year").PerCo.sum()
numer = veg_gc.groupby("Year").PerCo.sum()
(numer/denom).plot()
```

```python
#Graph of PerCo for GC all veg but not averaged
f, (ax1, ax2) = plt.subplots(nrows=2)
nonveg = ['BARE', 'DUFF', 'DOM', 'LI', 'LITR', 'ALG', 'ALGA', 'ALGB', 'ALGG','HAY', 'OPWA', 'H2O', 'MF',"WATER","LI","WRAC"]
veg_gc = data[~data.SPP.isin(nonveg)]
piv = veg_gc.pivot_table(values='PerCo', rows='Year', cols='SPP', aggfunc='sum')
denom = data.groupby("Year").PerCo.sum()
spwise = piv.transpose()
piv2 = spwise[spwise.max(axis=1)>600].transpose()
piv2.plot(ax=ax1)
ax1.set_title("Guadal Canal - Vegetation Percent Cover")

veg_ts = data2[~data2.SPP.isin(nonveg)]
piv = veg_ts.pivot_table(values='PerCo', rows='Year', cols='SPP', aggfunc='sum')
denom = data2.groupby("Year").PerCo.sum()
spwise = piv.transpose()
piv2 = spwise[spwise.max(axis=1)>350].transpose()
piv2.plot(ax=ax2)
ax2.set_title("Tubbs Setback Vegetation Percent Cover")
f.set_size_inches(6, 10)

plt.savefig("GCTS_veg_codes_all_years.png")
#(numer/denom).plot()
```

```python
def plot_species(data, species):
    sp = data[data.SPP == species]
    habitats = set(sp.Habitat)
    for habitat in habitats:
        df = sp[sp.Habitat == habitat]
        if len(df) == 0:
            continue
        perco = df.groupby('Date').PerCo
        perco.mean().plot(label=habitat)
    plt.title(species)
    plt.legend(loc=0)
```

```python
def plot_species_2(data, spp):
    habitats = set(data.Habitat)
    sp = data[data.SPP == spp]
    #ax = plt.subplot(len(habitats), 1, i)
    #plt.title(spp)

    for i, habitat in enumerate(habitats):
        ax = plt.subplot(len(habitats), 1, i)
        df = sp[sp.Habitat == habitat]
        if len(df) == 0:
            continue
        perco = df.groupby('Date').PerCo
        perco.mean().plot(label=habitat)
        plt.legend(loc='center right')

plot_species_2(data, 'SAPA')
```
